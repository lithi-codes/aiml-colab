```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    m = 0
    b = 0
    learning_rate = 0.01
    iterations = 10000
    n = len(x)

    for i in range(iterations):
        y_predicted = m * x + b

        md = -(2/n) * np.sum(x * (y - y_predicted))
        bd = -(2/n) * np.sum(y - y_predicted)

        m = m - learning_rate * md
        b = b - learning_rate * bd

    print(f"m = {m}, b = {b}")

    # Plot
    plt.scatter(x, y, color='red', marker='+', linewidth=3)
    plt.plot(x, y_predicted, color='green')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Gradient Descent Regression Line")
    plt.show()


x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

gradient_descent(x, y)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

def gradient_descent(x, y):
    m = 0
    b = 0
    iterations = 100000
    learning_rate = 0.0002
    n = len(x)

    prev_cost = float("inf")

    for i in range(iterations):
        y_predicted = m * x + b
        cost = np.mean((y - y_predicted) ** 2)

        md = -(2/n) * np.sum(x * (y - y_predicted))
        bd = -(2/n) * np.sum(y - y_predicted)

        m -= learning_rate * md
        b -= learning_rate * bd

        if abs(prev_cost - cost) < 1e-6:
            break

        prev_cost = cost

        if i % 10000 == 0:
            print(f"Iteration {i}: m={m}, b={b}, cost={cost}")

    return m, b


def predict_using_sklearn():
    df = pd.read_csv("test_scores.csv")
    r = LinearRegression()
    r.fit(df[['math']], df['cs'])
    return r.coef_[0], r.intercept_


if __name__ == "__main__":
    df = pd.read_csv("test_scores.csv")

    x = np.array(df['math'])
    y = np.array(df['cs'])

    m, b = gradient_descent(x, y)
    print(f"\nUsing Gradient Descent → m={m}, b={b}")

    m_sk, b_sk = predict_using_sklearn()
    print(f"Using sklearn → m={m_sk}, b={b_sk}")
```
