```python

# Single Variable Linear Regression

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Create sample dataset (instead of csv file)
data = {
    'area': [2600, 3000, 3200, 3600, 4000],
    'price': [550000, 565000, 610000, 680000, 725000]
}

df = pd.DataFrame(data)

# Plot data
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.show()

# Create and train model
reg = LinearRegression()
reg.fit(df[['area']], df['price'])

# Predict price for 3300 sq ft
print("Price for 3300 sq ft:", reg.predict([[3300]]))

# Show slope and intercept
print("Slope (m):", reg.coef_)
print("Intercept (b):", reg.intercept_)
```

```python
# Multiple Variable Linear Regression

data2 = {
    'area': [2600, 3000, 3200, 3600, 4000],
    'bedrooms': [3, 4, 3, 5, 4],
    'age': [20, 15, 18, 30, 8],
    'price': [550000, 565000, 610000, 680000, 725000]
}

df2 = pd.DataFrame(data2)

# Create model
reg2 = LinearRegression()
reg2.fit(df2[['area', 'bedrooms', 'age']], df2['price'])

# Predict
print("Price for 3000 sq ft, 3 bed, 40 age:",
      reg2.predict([[3000, 3, 40]]))

print("Price for 2500 sq ft, 4 bed, 5 age:",
      reg2.predict([[2500, 4, 5]]))

# Show coefficients
print("Coefficients:", reg2.coef_)
print("Intercept:", reg2.intercept_)
```

```python
# Hiring Salary Prediction

data3 = {
    'experience': [1, 2, 3, 5, 7, 10],
    'test_score': [8, 9, 6, 7, 10, 9],
    'interview_score': [6, 7, 7, 8, 9, 10],
    'salary': [50000, 60000, 65000, 80000, 95000, 120000]
}

df3 = pd.DataFrame(data3)

reg3 = LinearRegression()
reg3.fit(df3[['experience', 'test_score', 'interview_score']], df3['salary'])

# Predictions
print("Salary for 2 yr exp, 9 test, 6 interview:",
      reg3.predict([[2, 9, 6]]))

print("Salary for 12 yr exp, 10 test, 10 interview:",
      reg3.predict([[12, 10, 10]]))
```
